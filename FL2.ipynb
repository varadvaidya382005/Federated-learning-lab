```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("study_hours,attendance,internal_mar - study_hours,attendance,internal_mar (1).csv")

print(df.head())
print(df.info())
```

```python
# Fill numeric missing values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

print(df.isnull().sum())
```

```python
X = df[['study_hours', 'attendance', 'internal_marks']]
y = df['final_score']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
def create_clients(X, y, num_clients=3):
    client_data = {}

    data_size = len(X)
    shard_size = data_size // num_clients

    for i in range(num_clients):
        start = i * shard_size
        end = start + shard_size

        client_name = f"client_{i+1}"
        client_data[client_name] = (X[start:end], y.iloc[start:end])

    return client_data

clients = create_clients(X_train, y_train, num_clients=3)
```

```python
for client, data in clients.items():
    print(client, "Data Size:", len(data[0]))
```
